```python
import streamlit as st
import uuid
import joblib
import os
from pathlib import Path
from pipeline_helpers import ExtractAndPredict, CleanParagraphs, TitleGenerator
# Pipeline loads from pickled file
pipeline = joblib.load(Path("models/process_para_pipeline.pkl"))  

st.title("Paragraph Segmentation App")

uploaded_file = st.file_uploader("Choose a file", type=['docx','txt','pdf'], accept_multiple_files=False) 

if uploaded_file is not None:

    # Save file to upload folder
    file_id = uuid.uuid4().hex 
    file_path = Path(f"uploaded_files/{file_id}_{uploaded_file.name}")  
    with open(file_path,"wb") as f: 
        f.write(uploaded_file.getbuffer())

    # Run pipeline    
    html_path = Path(f"output/{file_id}_{uploaded_file.name}.html") 
    paragraphs = pipeline.transform([str(file_path)])

    # Display output file
    with open(html_path) as f:
        contents = f.read()
    st.components.v1.html(contents, width=700, height=800, scrolling=True)

    # Search 
    query = st.text_input("Search query:") 
    if query:
        results = search_paragraphs(query, paragraphs) 
        st.write("**Results:**")
        for par in results:
            st.write(par)
```
